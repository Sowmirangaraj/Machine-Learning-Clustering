```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import Birch
brc = Birch(threshold=0.5, branching_factor=50, n_clusters=3)
brc.fit(X)

```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised
```

```python
supervised['Cluster_group']=brc
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```
